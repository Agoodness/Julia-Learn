```julia
D = [i*j for i=1:5, j=1:5]
```

```julia
begin
    E = fill(0, 5, 5)
    for i = 1:5, j=1:5 #double loop
        E[i,j] = i*j
    end 
    E
end
```

```julia
begin
    F = fill(0, 5, 5)
    for i = 1:5
        for j = 1:5 # Nested loop
            F[i, j] = i*j
        end
    end 
    F
end
```
