We want to develop a function that accept a vector and find the cosine function of the vector

The cosine function = 1 - x²/2 + x^4/24 + x^6/720

```julia
function approxcos(x)
    outval = Array{Any}(undef, size(x))

    ii = 0
    for aa ∈ x
        y = aa^2/2 + aa^4/24 + aa^6/720 + aa^8/(56*720)
        ii += 1
        outval[ii] = y
    end
    return outval
end

x1 = 4*rand(10)

x2 = range(0., stop=4., step=0.01)

y1 = approxcos(x1) 

y2 = sin.(x2)
```
