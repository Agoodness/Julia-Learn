```julia
#This loop convert all the strings present in our data to zero (First letter of the string will be checked) 

rows, cols = size(EVDdata)
for j = 1:cols
    for i = 1:rows
        if !isdigit(string(EVDdata[i, j])[1])
            EVDdata[i, j] = 0
        end 
    end 
end 
```
